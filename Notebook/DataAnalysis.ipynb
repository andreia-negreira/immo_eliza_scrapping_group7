```python
df_localities = pd.read_excel("Data/Belgiesche_gemeenten_met_postcodes.xlsx")
df_localities = df_localities[['Postcode', 'Provincie']].drop_duplicates().reset_index(drop=True)
print(df_localities.head())
```

```python
df_localities['Gemeenschap'] = np.where(
    df_localities.Provincie.isin(['Vlaams-Brabant', 'Antwerpen', 'Limburg', 'West-Vlaanderen', 'Oost-Vlaanderen']),
    'Flanders',
    np.where(df_localities.Provincie == 'Brussels Hoofdstedelijk Gewest',
             'Brussels',
             'Wallonia'
    )
)
```

```python
df_localities
```

```python
# df_clean_data.drop(columns=['Provincie_x', 'index', 'Provincie_y', 'Gemeenschap'], inplace=True)
```

```python
df_clean_data = df_clean_data.merge(df_localities, on=['Postcode'], how='left')
df_clean_data
```

```python
# Read the Excel file
df_coordination = pd.read_csv('data/be.csv')
df_clean_data = df_clean_data.merge(df_coordination, on=['Locality'], how='left')
df_clean_data

```

```python


import folium
dict_data = df_clean_data.to_dict(orient='list')

belgium_map = folium.Map(location=[50.5039, 4.4699], zoom_start=8)
df = pd.DataFrame(dict_data)
for index, row in df.iterrows():
   if not pd.isna(row['Latitude']) and not pd.isna(row['Longitude']):
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Locality']}, Price: {row['Price']}",
        icon=folium.Icon(color='blue', icon='home')
    ).add_to(belgium_map)

belgium_map.save("belgium_map.html")



```
